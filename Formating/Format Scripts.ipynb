```python
def Avg_DialogueLine_length(Dialogue):
    avg = 0
    #print(type(Dialogue))
    #print(Dialogue)
    for line in Dialogue:
        avg += len(line)
    avg = float(avg / len(Dialogue))
    return avg
```

```python
def PreProssing(script,names, Key_words):
    '''Preprocessing the text'''
    text = re.split("[\.!?]",script)
    removables = []
    for x in range(0,len(text)):
        text[x] = text[x].lstrip()
        text[x] = text[x].strip()

    for x in range(0,len(text)):
        text[x] = re.sub(' +', ' ', text[x])

    text = list(filter(lambda x: x!= ' ' or x!= [], text))
    for x in range(0,len(text)):
        y = re.search('^[A-Z]{2}',text[x])
        temp = text[x].split()

        if(temp == [] or temp == ' '):
            text[x-1] += text[x]
            removables.append(text[x])
            continue
        #z = re.search('^[A-Z]{1}',text[x])
        if(y != None):
            pass
        else:
            #print(text[x])
            text[x-1] += ' '+ text[x]
            removables.append(text[x])
        if(len(temp) ==1):
            if(temp in Key_words):
                text[x] += ' ' + text[x+1]
                removables.append(text[x+1])
            else:
                text[x-1] += ' ' + text[x]
                removables.append(text[x])
    text = list(filter(lambda x: x not in removables,text))
    for x in range(0,len(text)):            
        y = re.search('^[A-Z]{2}',text[x])
        if(y):
            pass
        else:
            text[x-1] = text[x]
            removables.append(text[x])
    text = list(filter(lambda x: x not in removables,text))
    return text
    #DEbugging
```

```python
from collections import Counter
import re
def SuperScripting(script, Key_words,names, freq_caps, freq):
    words = Counter(Key_words)
    day_scenes = words['DAY']
    night_scenes = words['NIGHT']
    ext_scenes = words['EXT']
    interior_scenes = words['INT']
    fade_count = words['FADE']
    most_common = freq.most_common(3)
    most_important_character = freq_caps.most_common(2)
    descriptive_words = len(Key_words)
    
    text = PreProssing(script,names,Key_words)
    
    expostion_text = []
    Dialoguge_text = []
    misscilenous = []
    for line in text:
        tmp = line.split()
        #print(line)
        #print(tmp[0])
        if tmp[0] in names:
            Dialoguge_text.append(line)
        elif tmp[0] in Key_words:
            expostion_text.append(line)
        else:
            misscilenous.append(line)
    
    expostion_text += misscilenous
    try:
        dialogue_percentage = float(len(Dialoguge_text) / len(text))
    except:
        dialogue_percentage = None
    try:
        avg_length_of_dialogue = Avg_DialogueLine_length(Dialoguge_text)
    except:
        avg_length_of_dialogue = None
    try: 
        avg_length_of_exposition = Avg_DialogueLine_length(expostion_text)
    except:
        avg_length_of_exposition = None
    
    try: 
        exposition_percentage = float(len(expostion_text) / len(text))
    except:
        print(len(text))
        exposition_percentage = None
    lines_of_dialogue = len(Dialoguge_text)
    lines_of_exposition = len(expostion_text)
    
        
    
    return [words,day_scenes,night_scenes,ext_scenes,interior_scenes,fade_count,most_important_character,most_common,descriptive_words,dialogue_percentage, exposition_percentage,lines_of_dialogue,lines_of_exposition,avg_length_of_dialogue,avg_length_of_exposition,expostion_text,Dialoguge_text,misscilenous]

# capture dialogue density
# number of characters lines/ runtime
# look at the switching between dialogue / runtime
# sentiment analysis count natural word frequency for lines
# plain dialogue intesity metric
# weighted dialogue intesity metric
# text to image using descriptive exposition scenes
# natural frequency score (compare the natural frequency score) avg, median, variance 
# combination of the word
```
